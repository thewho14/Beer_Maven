```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans()
```

```python
df.groupby('username').count().sort_values('rating',ascending=False).head(10)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
user_nums = le.fit_transform(df.username.values)
user_labs = le.classes_
type_nums = le.fit_transform(df.type.values)
type_labs = le.classes_
group_nums = le.fit_transform(df.group.values)
group_labs = le.classes_
```

```python
numbed = df.copy()
numbed.loc[:,'username'] = user_nums
numbed.loc[:,'type'] = type_nums
numbed.loc[:,'group'] = group_nums
numbed.head()
```

```python
kfc = numbed[['username','type','group','rating']]
kmeans.fit_predict(kfc)
kmeans.inertia_
```
